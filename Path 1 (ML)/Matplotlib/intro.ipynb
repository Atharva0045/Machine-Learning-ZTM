```python
X = [1,2,3,4]
Y = [11,22,33,44]
plt.plot(X,Y)
plt.show()
```

```python
# 1st Method: Pyplot API
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# 2nd Method: Same as Pyplot API
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(X,Y)
plt.show()
```

```python
# 3rd Method: Object Oriented API (RECOMMENDED METHOD)
fig, ax = plt.subplots()
ax.plot(X,Y)
plt.show()
```

Creating plots from Dict

```python
nut_prices = {'Almond': 12, 
              'Peanut': 10,
              'Cashew': 14}

fig, ax = plt.subplots()

ax.set(title='Nut store Prices',
       xlabel='Types of butter',
       ylabel='Price')
ax.bar(nut_prices.keys(), nut_prices.values()) # vertical bar plot
```

```python
fig, ax = plt.subplots()

ax.set(title='Nut store Prices',
       xlabel='Price',
       ylabel='Types of butter')
ax.barh(list(nut_prices.keys()), list(nut_prices.values()))
```

Example from random numbers for subplots

```python
heart_dataframe = pd.read_csv('./heart-disease.csv')
```

```python
heart_dataframe
```

```python
heart_dataframe['age'].plot.hist(bins=20)
```

```python
heart_dataframe.head()
```

```python
age_over_50 = heart_dataframe[heart_dataframe['age'] > 50]
```

```python
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=age_over_50['age'],
                     y=age_over_50['chol'],
                     c=age_over_50['target'])

ax.set(title='Heart Disease Cholesterol level',
       xlabel='Age',
       ylabel='Cholesterol')

ax.legend(*scatter.legend_elements(), title='Target')

ax.axhline(age_over_50['chol'].mean(),
           linestyle='--')
```

```python
age_over_50
```

Subplot of Chol, age and thalach

```python
plt.style.use('seaborn-v0_8-whitegrid')

fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               sharex=True,
                               figsize=(10, 10))

scatter = ax0.scatter(x=age_over_50['age'],
                      y=age_over_50['chol'],
                      c=age_over_50['target'],
                      cmap='winter')

ax0.set(title='Cholesterol Levels',
        ylabel='Cholesterol',
        xlim=[50,80],
        ylim=[100, 600])

ax0.legend(*scatter.legend_elements(), title='Target')

ax0.axhline(age_over_50['chol'].mean(),
            linestyle='--',
            color='red')

scatter2 = ax1.scatter(x=age_over_50['age'],
                      y=age_over_50['thalach'],
                      c=age_over_50['target'],
                      cmap='winter')

ax1.set(title='Thalach Levels',
        xlabel='Age',
        ylabel='Max Heart Rate',
        xlim=[50,80],
        ylim=[60, 200])

ax1.legend(*scatter2.legend_elements(), title='Target')

ax1.axhline(age_over_50['thalach'].mean(),
            linestyle='--',
            color='red')


fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
```
