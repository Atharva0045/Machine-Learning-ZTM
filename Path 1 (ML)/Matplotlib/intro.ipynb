```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Intro

```python
plt.plot()
plt.show()
```

```python
X = [1,2,3,4]
Y = [11,22,33,44]
plt.plot(X,Y)
plt.show()
```

```python
# 1st Method: Pyplot API
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# 2nd Method: Same as Pyplot API
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(X,Y)
plt.show()
```

```python
# 3rd Method: Object Oriented API (RECOMMENDED METHOD)
fig, ax = plt.subplots()
ax.plot(X,Y)
plt.show()
```
